```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import os
from zipfile import ZipFile
import fiona
import geopandas as gpd
import folium
```

Parámetros de entrada

```python
file_name = 'Cultivo_1'
fecha_ini = '2015-08-24'
fecha_fin = '2016-02-11'
cct = 90
```

Rutas y fechas de análisis

```python
ruta_base = './00_Informacion/'

ruta_cultivos = ruta_base + 'Cultivos/'
ruta_download = ruta_base + 'download/'
ruta_wrs      = ruta_base + 'WRS2_descending_0/'

fechas_analisis = pd.date_range(fecha_ini, fecha_fin)
```

Paths y Rows de satélite

```python
wrs_file = ruta_wrs + 'WRS2_descending.shp'
wrs = gpd.GeoDataFrame.from_file(wrs_file)
```

```python
wrs.head()
```

From kmz to kml

```python
fiona.drvsupport.supported_drivers['kml'] = 'rw'
fiona.drvsupport.supported_drivers['KML'] = 'rw'

kmz = ZipFile(ruta_cultivos + file_name + '.kmz', 'r')
kmz.extract('doc.kml', ruta_cultivos)

cultivo = gpd.read_file(os.path.join(ruta_cultivos, 'doc.kml'))
```

Intersección

```python
wrs_intersection = wrs[wrs.intersects(cultivo.geometry[0])]
paths, rows = wrs_intersection['PATH'].values, wrs_intersection['ROW'].values
```

Mapa interactivo

```python
# Centro del mapa
xy = np.asarray(cultivo.centroid[0].xy).squeeze()
center = list(xy[::-1])

# Define un zoom inicial
zoom = 6

# Creación del mapa
m = folium.Map(location=center, zoom_start=zoom, control_scale=True)

# Plotea el shape del cultivo
m.add_child(folium.GeoJson(cultivo.__geo_interface__, name='Area of Study', style_function=lambda x: {'color': 'red', 'alpha': 0}))

# Plotea los recuadros que intersecan el área
for i, row in wrs_intersection.iterrows():
    # Crea un string para el nombre que contiene el path y row del polígono
    name = 'path: %03d, row: %03d' % (row.PATH, row.ROW)
    
    g = folium.GeoJson(row.geometry.__geo_interface__, name=name)
    g.add_child(folium.Popup(name))
    g.add_to(m)

folium.LayerControl().add_to(m)
m.save(ruta_base + 'mapa/wrs.html')
m
```

Carga lista de archivos en Amazon

```python
# s3_scenes = pd.read_csv(ruta_base + 'data/scene_list.gz', compression='gzip')
s3_scenes = pd.read_csv('http://landsat-pds.s3.amazonaws.com/c1/L8/scene_list.gz', compression='gzip')
```

```python
s3_scenes.head()
```

Descarga datos de acuerdo a la información de entrada

```python
bulk_list = pd.DataFrame(columns=s3_scenes.columns)

for path, row in zip(paths, rows):

    print('Path:',path, 'Row:', row)

    scenes = s3_scenes[(s3_scenes.path == path) & (s3_scenes.row == row) & 
                       (s3_scenes.cloudCover <= cct) & 
                       (~s3_scenes.productId.str.contains('_T2')) &
                       (~s3_scenes.productId.str.contains('_RT'))]
    print(u' Fueron encontradas {} imágenes'.format(len(scenes)))

    fechas_scenes = pd.to_datetime(scenes.acquisitionDate.values).strftime('%Y-%m-%d')
    
    tmp = []
    for i in range(len(fechas_scenes)):
        if fechas_scenes[i] in fechas_analisis:
            tmp.append(i)
    bulk_list = bulk_list.append(scenes.iloc[tmp])
    
print(u'\nEn el período de tiempo ingresado hay {} imágenes disponibles'.format(len(bulk_list)))
    
print(u'\nInicia el proceso de descarga')
    
for k in range(len(bulk_list)):
    
    print(' Descargando {} de {} imágenes'.format(k+1, len(bulk_list)))
    
#     URL = bulk_list.download_url.values[k][:-10]
#     productID = bulk_list.productId.values[k]
#     os.system('wget -P ' + ruta_download + ' ' + URL + productID + '_B4.TIF')
#     os.system('wget -P ' + ruta_download + ' ' + URL + productID + '_B5.TIF')
```

Funciones

```python
def Reproject_Raster(ruta_in, ruta_out, dst_crs='EPSG:4326'):

    import numpy as np
    import rasterio
    from rasterio.warp import calculate_default_transform, reproject, Resampling

    with rasterio.open(ruta_in) as src:
        transform, width, height = calculate_default_transform(
            src.crs, dst_crs, src.width, src.height, *src.bounds)
        kwargs = src.meta.copy()
        kwargs.update({
            'crs': dst_crs,
            'transform': transform,
            'width': width,
            'height': height
        })

        with rasterio.open(ruta_out, 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=dst_crs,
                    resampling=Resampling.nearest)
                
def draw_grid(minlon, maxlon, deltalon, minlat, maxlat, deltalat):
    
    xticks = np.arange(minlon, maxlon, deltalon)
    yticks = np.arange(minlat, maxlat, deltalat)
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, color='k', linestyle='--', linewidth=0.4)
    gl.xlocator   = mticker.FixedLocator(xticks)
    gl.ylocator   = mticker.FixedLocator(yticks)
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    gl.top_labels    = False
    gl.right_labels  = False
    gl.xlabels_top   = False
    gl.ylabels_right = False
```

Cálculo del NDVI

```python
fechas_imagenes = pd.to_datetime(pd.to_datetime(sorted(bulk_list.acquisitionDate.values)).strftime('%Y-%m-%d'))

serie_NDVI = []

for i_fecha in fechas_imagenes:
    
    print(i_fecha.strftime('%Y-%m-%d'))
    
#     i_fecha = fechas_imagenes[4]

    file_banda4 = glob.glob(ruta_download + '*' + i_fecha.strftime('%Y%m%d') + '*B4.TIF')[0]
    file_banda5 = glob.glob(ruta_download + '*' + i_fecha.strftime('%Y%m%d') + '*B5.TIF')[0]

    Reproject_Raster(file_banda4, ruta_download + 'tmp_B4_WGS84.tif')
    Reproject_Raster(file_banda5, ruta_download + 'tmp_B5_WGS84.tif')

    banda4 = rasterio.open(ruta_download + 'tmp_B4_WGS84.tif')
    banda5 = rasterio.open(ruta_download + 'tmp_B5_WGS84.tif')
    
    if i_fecha == fechas_imagenes[0]:

        print('Banda 4')
        plt.figure()
        plt.imshow(banda4.read(1))
        plt.colorbar()
        plt.show()

        print('Banda 5')
        plt.figure()
        plt.imshow(banda5.read(1))
        plt.colorbar()
        plt.show()

    cut_banda4, out_transform_banda4 = rasterio.mask.mask(banda4, [cultivo.geometry[0]], crop=True)
    cut_banda5, out_transform_banda5 = rasterio.mask.mask(banda5, [cultivo.geometry[0]], crop=True)

    NDVI = ((cut_banda5 - cut_banda4) / (cut_banda5 + cut_banda4))[0]

    lon = rasterio.transform.xy(out_transform_banda4, range(NDVI.shape[1]), range(NDVI.shape[1]))[0]
    lat = rasterio.transform.xy(out_transform_banda4, range(NDVI.shape[0]), range(NDVI.shape[0]))[1]
    
    serie_NDVI.append(NDVI)
    
serie_NDVI = np.array(serie_NDVI)
```
